```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df = pd.read_csv('Social_Network_Ads.csv')

print(df.to_string())
```

```python
print("Number of missing data in every columns: ")
df.isna().sum()
```

```python
print("Number of duplicated data in every columns: ",df.duplicated().sum())
```

```python
df[['Age','EstimatedSalary']].describe().T
```

```python
df['Gender'] = df['Gender'].replace(['Male'], 1)
df['Gender'] = df['Gender'].replace(['Female'], 0)
```

```python
print(df.to_string())
```

```python
import numpy as np

df_x = df[['Gender', 'Age', 'EstimatedSalary']]
df_y = df[['Purchased']]
X = df_x.values
y = df_y.values
print(X)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler() 
X_scaled = ss.fit_transform(X) 
```

```python
X_scaled_train, X_scaled_test, y_scaled_train, y_scaled_test = train_test_split(X_scaled, y,train_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(X_scaled_train, y_scaled_train)
```

```python
y_pred = log.predict(X_scaled_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_scaled_test, y_pred) * 100
print(acc)
```

```python
print(f"h(x) = {log.intercept_} + [{log.coef_[0][0]}] X1 + [{log.coef_[0][1]}] X2 + [{log.coef_[0][2]}] X3")
```
