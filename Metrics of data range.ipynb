### 4.1 Нарисуйте scatter plot количества лошадинных сил против рабочего объема (displacement). Что вы думаете об их связи между собой?

```python
x = data['horsepower']
y = data['displacement']

plt.scatter(x, y, s=100, c='red', marker='o')
plt.title("Histogram of Mpg size")
plt.xlabel("Horsepower")
plt.ylabel("Displacement")

# В данном случае взаимоотношение колличества лошадинных сил против рабочего объема имеет высокую степень связи. Можно заметить что
 #при увеличении логадинных сил, рабочий объем ростет. Прямая корреляция
```

### 4.2 Нарисуйте scatter plot  рабочего объема против mpg. Что вы думаете об их связи между собой?

```python
x = data['horsepower']
y = data['mpg']

plt.scatter(x, y, s=100, c='red', marker='o')
plt.title("Histogram of Mpg size")
plt.xlabel("Horsepower")
plt.ylabel("Mpg")

# При увеличении рабочего объема, миля на галлону уменьшается 
# что миля на галлон уменьшается при увеличении рабочего объема, имеет смысл с точки зрения того, что автомобили 
# с большими двигателями часто требуют  больше топлива на единицу пройденного расстояния, что сказывается на их экономичности

```

### 4.3 Нарисуйте scatter plot  рабочего объема против mpg и покрасьте точки в зависимости от страны производства автомобиля (hue=...).  Что вы думаете?

```python
x = data['displacement']
y = data['mpg']
origin = data['origin']

plt.figure(figsize=(10,8))
sns.scatterplot(data=data, x='displacement', y='mpg', hue='origin')

plt.title("Scatter Plot: Displacement vs. MPG (colored by Origin)")
plt.xlabel("Dsiplacement")
plt.ylabel("Mpg")
plt.legend(title='Origin')
plt.grid(True)
plt.show()


# Япония показывает самый высокий уровень миль за галлон при низком рабочем объеме двигателя, это может быть связано с тем, 
# что японские производители автомобилей, такие как Toyota и Honda, часто известны своими эффективными и 
# экономичными двигателями. Они сфокусированы на разработке технологий, которые улучшают топливную эффективность и снижают выбросы
# В Европе также автомобили, которые обычно имеют более маленькие двигатели с высокой мощностью
#в то время как США показывает самые высокий рабочий объем в то время как мили за галлону не достигает больше чем 40 м/г
# США известны своими крупными автомобилями, включая пикапы и внедорожники, которые обычно имеют более мощные двигатели 
# и более низкую экономичность топлива из-за их размера и веса.
```
