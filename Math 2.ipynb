```python
#5. Найдите производную функции f(x) и создайте новую функцию на ее основе как f_diff(x)
```

```python
import sympy as sp
x = sp.symbols ('x')

sp.diff((x ** 6) + (10 * (x ** 3)) - 1000)

x_list = []
y_list = []

for i in range (-10, 11):
    x_list.append(i)
    y_list.append(i**5)
print(x_list)
print(y_list)

plt.plot(x_list, label = "Function")
y_list_derivative = []
for i in range (-10,11):
    y_list_derivative.append(5 * i ** 4)
print (y_list_derivative)

plt.plot(x_list, y_list_derivative, label = "Derivative")
plt.show()

```
